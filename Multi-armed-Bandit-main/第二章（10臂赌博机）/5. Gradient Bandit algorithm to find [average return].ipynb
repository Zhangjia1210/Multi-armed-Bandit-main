```python
import numpy as np
import random
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
threshold = 100


step = 1000
runs = 2000
sigma = 1  # 10-armed testbed 
miu_lst = [0.2, -0.85, 1.55, 0.3, 1.2, -1.5, -0.2, -1.0, 0.9, -0.6]  # 10-armed testbed 


def action_reward(action):  # 10-armed testbed
    reward = miu_lst[action]+np.random.normal(0, scale=sigma)
    return reward


def random_index(rate):  # 
    # """"""
    # ratelist<int>
    # 
    for i in range(0, len(rate)):
        rate[i] = int(1000*rate[i])
    start = 0
    index = 0
    randnum = random.randint(1, sum(rate))
    for index, scope in enumerate(rate):
        start += scope
        if randnum <= start:
            break
    return index


def max_index(lst):  # 
    index_lst = []
    max_n = max(lst)
    for i in range(len(lst)):
        if lst[i] == max_n:
            index_lst.append(i)
    max_index = random.sample(index_lst, 1)[0]
    return max_index


def Gradient_Bandit_withbase(alpha, step, runs):  #  preference
    Q_runs_lst = [[] for m in range(runs)]  # runsstepQ
    Q_runs_lst_average = [0 for n in range(step)]  #  Q_runs_lst runsstepQ
    for run_times in range(0, runs):
        Q = [4 for i in range(10)]  # (greedaverage reward)Q=+4
        N = [0 for j in range(10)]  # 
        H = [0 for k in range(10)]  # P
        H_e = [0 for l in range(10)]  # P
        P = [0 for m in range(10)]  # 
        k = 1
        while k <= step:
            if sum(H) == 0:
                action_id = random.choice([i for i in range(10)])
            else:
                for i in range(0, len(H_e)):
                    H_e[i] = 2.71828 ** H[i]
                for i in range(0, len(P)):
                    P[i] = H_e[i]/sum(H_e)
                action_id = max_index(P)  # P
            N[action_id] = N[action_id] + 1
            act_reward = action_reward(action_id)  # 
            Q_runs_lst[run_times].append(act_reward)
            Q[action_id] = Q[action_id] + (1/N[action_id])*(act_reward - Q[action_id])  # 
            for i in range(0, len(H)):
                if i == action_id:
                    H[i] = H[i] + alpha * (act_reward - Q[i])*(1-P[i])  # H
                    if H[i] > threshold:  # H
                        H[i] = threshold
                    if H[i] < -threshold:
                        H[i] = -threshold
                else:
                    H[i] = H[i] - alpha * (act_reward - Q[i]) * P[i]
                    if H[i] > threshold:  # H
                        H[i] = threshold
                    if H[i] < -threshold:
                        H[i] = -threshold
            k += 1
    for i in range(0, step):
        for j in range(0, runs):
            Q_runs_lst_average[i] = Q_runs_lst_average[i] + Q_runs_lst[j][i]
        Q_runs_lst_average[i] = Q_runs_lst_average[i]/runs
    return Q_runs_lst_average


def Gradient_Bandit_withoutbase(alpha, step, runs):  #  preference
    Q_runs_lst = [[] for m in range(runs)]  # runsstepQ
    Q_runs_lst_average = [0 for n in range(step)]  #  Q_runs_lst runsstepQ
    for run_times in range(0, runs):
        Q = [4 for i in range(10)]  # (greedaverage reward)Q=+4
        N = [0 for j in range(10)]  # 
        H = [0 for k in range(10)]  # P
        H_e = [0 for l in range(10)]  # P
        P = [0 for m in range(10)]  # 
        k = 1
        while k <= step:
            if sum(H) == 0:
                action_id = random.choice([i for i in range(10)])
            else:
                for i in range(0, len(H_e)):
                    H_e[i] = 2.71828 ** H[i]
                for i in range(0, len(P)):
                    P[i] = H_e[i]/sum(H_e)
                action_id = max_index(P)  # P
            N[action_id] = N[action_id] + 1
            act_reward = action_reward(action_id)  # 
            Q_runs_lst[run_times].append(act_reward)
            Q[action_id] = Q[action_id] + (1/N[action_id])*(act_reward - Q[action_id])  # 
            for i in range(0, len(H)):
                if i == action_id:
                    H[i] = H[i] + alpha * act_reward *(1-P[i])  # H
                    if H[i] > threshold:  # H
                        H[i] = threshold
                    if H[i] < -threshold:
                        H[i] = -threshold
                else:
                    H[i] = H[i] - alpha * act_reward * P[i]
                    if H[i] > threshold:  # H
                        H[i] = threshold
                    if H[i] < -threshold:
                        H[i] = -threshold
            k += 1
    for i in range(0, step):
        for j in range(0, runs):
            Q_runs_lst_average[i] = Q_runs_lst_average[i] + Q_runs_lst[j][i]
        Q_runs_lst_average[i] = Q_runs_lst_average[i]/runs
    return Q_runs_lst_average


_010_GB_withbase = Gradient_Bandit_withbase(0.1, step, runs)
_040_GB_withbase = Gradient_Bandit_withbase(0.9, step, runs)
# _010_GB_withoutbase = Gradient_Bandit_withoutbase(0.1, step, runs)
# _040_GB_withoutbase = Gradient_Bandit_withoutbase(0.4, step, runs)


x = np.arange(step)
plt.plot(x, _010_GB_withbase, color='red')
plt.plot(x, _040_GB_withbase, color='blue')
# plt.plot(x, _010_GB_withoutbase, color='blue')
# plt.plot(x, _040_GB_withoutbase, color='royalblue')
plt.legend(["alpha=0.1 with baseline, ", "alpha=0.9 with baseline"])  # "alpha=0.1 without baseline", "alpha=0.4 without baseline"
plt.grid(color='gray', linestyle=':')
plt.xlabel("Step", fontweight='bold', fontsize=14, labelpad=2)
plt.ylabel('Average reward', fontweight='bold', fontsize=14, labelpad=2)
plt.show()
```
